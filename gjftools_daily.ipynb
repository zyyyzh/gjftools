### Freeze all atoms except one (or a few)

useful when calculating fragment interaction

```python
def freeze_atom_gjf(gjffile, model_file, result_file, no_freeze_atom=-1):
    if isinstance(no_freeze_atom, int):
        no_freeze_atom = [no_freeze_atom]
    gjf = gjfdata()
    gjf.get_body_from_gjf(gjffile)
    gjf.get_model_file(model_file)
    type_list = ['-1' for i in range(len(gjf))]
    for index in no_freeze_atom:
        type_list[index] = '0'
    gjf.atom_type = type_list
    gjf.write_gjf(result_file, need_chk=False)
```

```python
gjf_dir = '/home/yzh/g09/structures/'
model_file = '/home/yzh/g09/models/ZCH/analysis/Ru2L3-M062X-frag-opt-model.gjf'
result_dir = '/home/yzh/g09'

for file in os.listdir(gjf_dir):
    if file.endswith('.gjf'):
        gjf_file = os.path.join(gjf_dir, file)
        result_file = os.path.join(result_dir, os.path.basename(gjf_file))
        freeze_atom_gjf(gjf_file, model_file, result_file)
```
